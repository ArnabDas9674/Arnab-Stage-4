```python
# Group the original DataFrame by 'Country' and calculate the mean of 'Value'
df = i1.groupby('Country')['Value'].mean().reset_index()

# Convert 'Country' column to categorical data type
df['Country'] = pd.Categorical(df['Country'])

# Convert categorical codes to integers
df['Country'] = df['Country'].cat.codes

# Create a boxplot of the 'Value' column
plt.boxplot(x = 'Value',data = df, widths = 800)
plt.show()
```
